```python
import csv
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def get_newest_file_name():
    dir_list = os.listdir()
    filtered_list = [k for k in dir_list if 'cs_scrape_' in k]
    if len(filtered_list) > 0:
        filtered_list.sort(reverse=True)
        return filtered_list[0]
    return None
```

```python
df = pd.read_csv(get_newest_file_name())
df
```

```python
servers = df["server"].value_counts()
servers.plot(kind='bar', color='blue', edgecolor='black', figsize=(17,8))
plt.title('Server Location Counts')
plt.xlabel('Server Location')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
cheater_games = df["cheater_names_str"].dropna()
non_cheater_games = df[df["cheater_names_str"].isna()]
print(cheater_games.count())
print(len(non_cheater_games))
plt.figure(figsize=(8,8))
plt.bar(["Cheater games","Non cheater games"], [cheater_games.count(),len(non_cheater_games)])
plt.ylabel("Game count")
plt.show()
```

```python
maps = df["map"].value_counts()
maps.plot(kind='bar', color='blue', edgecolor='black', figsize=(17,8))
plt.title('Map distributions')
plt.xlabel('Maps')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
